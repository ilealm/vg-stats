### 4. What are the top 20 highest grossing games?

```python
top_twenty = df_games[['Name','Global_Sales', 'NA_Sales']].sort_values('Global_Sales', ascending = False).nlargest(20, 'Global_Sales')
top_twenty
```

### 5. For North American video games sales, whats is the median?

```python
NA_median = df_games['NA_Sales'].median()
NA_median
```

### 5.b. Show 10 games sorrunding the NA median.

```python
games_NA_sorted = df_games.sort_values('NA_Sales', ascending = False)

NA_median_index = int(games_NA_sorted.shape[0]/2)
# NA_subset = df_games[df_games['NA_Sales'] < NA_median][['Name','NA_Sales']].sort_values('NA_Sales', ascending = False).nlargest(20, 'NA_Sales')
NA_subset = games_NA_sorted[NA_median_index-5:NA_median_index+5]
NA_subset
```

### 6.For the top selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
top_game = top_twenty.iloc[0]
NA_stddev = df_games['NA_Sales'].std()
NA_mean = df_games['NA_Sales'].mean()
(top_game['NA_Sales']-NA_mean)/NA_stddev
```

### 7. The nintendo WII seems to have outdone itself with games. How does its average number of sales compare with all of other plattforms?

```python
avg_by_platform = df_games[['Platform', 'Global_Sales']].groupby('Platform').mean().sort_values('Global_Sales', ascending = False)
# position of Wii sales
avg_by_platform.index.get_loc('Wii')+1
```

### 8. Come up with 3 more questions that can be answered with the data set

#### 8.1 What were the average sales by platform in 2010 and 2011 combined?

```python
df_games[df_games['Year'].isin([2010,2011])][['Platform', 'Global_Sales']].groupby('Platform').mean().sort_values('Global_Sales', ascending = False)
```

#### 8.2 What percentage of sales come from NA for each year?

```python
sales_by_year = df_games[['Year', 'NA_Sales', 'Global_Sales']].groupby('Year').sum()
sales_by_year['PCT'] = 100 * sales_by_year['NA_Sales'] / sales_by_year['Global_Sales']
sales_by_year
```

#### Top 10 genre that grew the most in 2015

```python
df_games['Previous_Year'] = df_games['Year'] - 1
games_sales = df_games[df_games['Year'].isin([2014,2015])][['Genre', 'Global_Sales', 'Year', 'Previous_Year']]
games_sales_YoY = games_sales.merge(games_sales, left_on = ['Genre', 'Year'], right_on = ['Genre', 'Previous_Year']).groupby('Genre').sum()
games_sales_YoY['Growth_PCT'] = 100*(games_sales_YoY['Global_Sales_x']-games_sales_YoY['Global_Sales_y'])/games_sales_YoY['Global_Sales_y']
games_sales_YoY.sort_values('Growth_PCT')[['Growth_PCT']]
# Name	Platform	Year	Genre	Publisher	NA_Sales	EU_Sales	JP_Sales	Other_Sales	Global_Sale
```
